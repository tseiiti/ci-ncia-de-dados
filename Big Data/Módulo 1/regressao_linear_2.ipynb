```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.linear_model import LinearRegression
```

```python
base_custo_diaria = pd.read_csv('regressao_linear_2.csv', sep = ';')
```

```python
base_custo_diaria.head()
```

```python
base_custo_diaria.describe()
```

```python
base_custo_diaria.info()
```

```python
base_custo_diaria.isna().sum()
```

```python
#base_custo_diaria.groupby(['qtd_pessoas']).describe()
```

```python
x_base = base_custo_diaria.iloc[:, 0].values
y_base = base_custo_diaria.iloc[:, 1].values
```

```python
np.corrcoef(x_base, y_base)
```

```python
x_shape = x_base.reshape(-1, 1)
x_shape[:5]
```

```python
lr = LinearRegression()
lr.fit(x_shape, y_base)
```

```python
# constante alpha (a)
lr.intercept_
```

```python
# constante beta (b)
lr.coef_
```

```python
previsoes = lr.predict(x_shape)
previsoes[:5]
```

```python
gf = px.scatter(x = x_base, y = y_base)
gf.add_scatter(x = x_base, y = previsoes, name = 'previsões')
gf.show()
```

```python
# previsão pelo método:
lr.predict([[57]])
```

```python
# diferença entre base e predito
i = 25
y_base[np.where(x_base == i)[0][0]] - lr.predict([[i]])[0]
```

```python
# score
lr.score(x_shape, y_base)
```

```python
# erro absoluto
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_absolute_error(y_base, previsoes)
```
