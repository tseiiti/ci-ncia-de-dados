```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.linear_model import LinearRegression
```

```python
base_custo_diaria = pd.read_csv('trabalho_pratico.csv', sep = ';')
```

```python
# 8. Qual o valor de correlação entre as variáveis qtd_pessoas e valor da diária?
x_base = base_custo_diaria.iloc[:, 0].values
y_base = base_custo_diaria.iloc[:, 1].values
np.corrcoef(x_base, y_base)
```

```python
# 9. Qual o valor retornado intercept da previsão?
x_shape = x_base.reshape(-1, 1)

lr = LinearRegression()
lr.fit(x_shape, y_base)

lr.intercept_
```

```python
# 10. Qual o valor da previsão realizada pelo modelo de regressão para a qtd_pessoas = 20 ?
qtd_pressoas = 20
lr.intercept_ + lr.coef_ * qtd_pressoas
```

```python
# 11. Qual o valor da previsão realizada pelo modelo de regressão para a qtd_pessoas = 57 ?
lr.predict([[57]])
```

```python
# 12. Qual a difença entre o valor predito e o valor real do valor da diária para a quantidade de pessoas = 25
qtd_pressoas = 25
y_base[np.where(x_base == qtd_pressoas)[0][0]] - lr.predict([[qtd_pressoas]])[0]
```

```python
# 13. Qual foi o score obtido pelo algoritmo de regressão?
lr.score(x_shape, y_base)
```

```python
# 14. Qual o valor obtido para mean_absolute_error?
from sklearn.metrics import mean_squared_error, mean_absolute_error
previsoes = lr.predict(x_shape)
mean_absolute_error(y_base, previsoes)
```

```python
gf = px.scatter(x = x_base, y = y_base)
gf.add_scatter(x = x_base, y = previsoes, name = 'previsões')
gf.show()
```
