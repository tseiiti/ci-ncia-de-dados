```python
df = pd.read_csv('associacao.csv', sep = ';', encoding= 'latin-1')
```

```python
df
```

```python
lst = []
for i in range(len(df)):
  lst.append([str(df.values[i, j]) for j in range(4)])
# lst
```

```python
from apyori import apriori
```

```python
# list(apriori(lst))
# list(apriori(lst, min_support = 0.2, min_confidence = 0.8))
result = list(apriori(lst, min_support = 0.3, min_confidence = 0.8, min_lift = 2))
# result = list(apriori(lst, min_support = 0.4, min_confidence = 0.8, min_lift = 1))
result
```

min_support: frequencia mínima do item de aparecer nos registros analisados
min_confidence: frequencia mínima da combinaçào de aparecer no registros com suporte mínimo
min_lift: frequencia mínima do item B (items_base) ser associado ao(s) item(s) A (items_add)
items=frozenset: retorna os itens sets avaliados
OrderedStatistic: mostra as regras criadas

```python
a = []
b = []
s = []
c = []
l = []
for res in result:
  for r in res[2]:
    a.append(list(r[0]))
    b.append(list(r[1]))
    s.append(res[1])
    c.append(r[2])
    l.append(r[3])

df = pd.DataFrame({'A': a, 'B': b, 'Suporte': s, 'Confiança': c, 'Lift': l})
df.sort_values(by = 'Lift', ascending = False)
```
