```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score
```

```python
df_diabetes = pd.read_csv('arvore_decisao.csv')
```

```python
df_diabetes.head()
df_diabetes.info()
df_diabetes.describe()
df_diabetes.isna().sum()
```

```python
# verifica que a quantidade de não diabéticos (Outcome = 0) é maior
df_diabetes.groupby(['Outcome']).describe()
```

```python
# n_splits = 1 => 1 divisão de conjuntos?
# test_size = 0.1 => teste com 10% da base?
def classificador(x, y, max_depth = 0):
  validator = StratifiedShuffleSplit(n_splits = 1, test_size = 0.1, random_state = 0)
  for treino_id, teste_id in validator.split(x, y):
    x_train, x_test = x[treino_id], x[teste_id]
    y_train, y_test = y[treino_id], y[teste_id]

  # treino
  if max_depth == 0:
    tc = tree.DecisionTreeClassifier(random_state = 0)
  else:
    # com poda da árvore
    tc = tree.DecisionTreeClassifier(max_depth = max_depth, random_state = 0)
  arvore = tc.fit(x_train, y_train)

  # teste => validação do treino
  y_pred = arvore.predict(x_test)

  return x_train, x_test, y_train, y_test, tc, y_pred
```

```python
# separa o campo Outcome (indicador de diabetes) da tabela. Seria o gabarito para os testes?
x = df_diabetes.drop('Outcome', axis = 1).values
y = df_diabetes['Outcome'].values

x_train, x_test, y_train, y_test, tc, y_pred = classificador(x, y, 0)

# validando acurácia (n. acertos y_pred / y_test)
accuracy_score(y_test, y_pred)
```

```python
 plt.figure(figsize = (100, 50))
 tree.plot_tree(tc, filled = True, fontsize = 16)
 plt.savefig('arvore.png')
 plt.close()
```
