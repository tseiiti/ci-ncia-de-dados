```python
# list(apriori(lst))
# list(apriori(lst, min_support = 0.2, min_confidence = 0.8))
# result = list(apriori(lst, min_support = 0.3, min_confidence = 0.8, min_lift = 2))
result = list(apriori(lst, min_support = 0.4, min_confidence = 0.8, min_lift = 1))
# result
```

- min_support: frequencia mínima do item de aparecer nos registros analisados
- min_confidence: frequencia mínima da combinação de aparecer no registros com suporte mínimo
- min_lift: frequencia mínima do item B (items_base) ser associado ao(s) item(s) A (items_add)

# Apresentação do resultado

```python
result = list(apriori(lst, min_support = 0.4, min_confidence = 0.8, min_lift = 1))

item_a = []
item_b = []
suport = []
confia = []
lift = []
for relation_record in result:
  for orddered_statistic in relation_record[2]:
    if 'nan' not in orddered_statistic[0] and 'nan' not in orddered_statistic[1]:
      item_a.append(list(orddered_statistic[0]))
      item_b.append(list(orddered_statistic[1]))
      suport.append(relation_record[1])
      confia.append(orddered_statistic[2])
      lift.append(orddered_statistic[3])

df = pd.DataFrame({'A': item_a, 'B': item_b, 'Suporte': suport, 'Confiança': confia, 'Lift': lift})
df.sort_values(by = ['Lift', 'Confiança', 'Suporte'], ascending = False)
```
